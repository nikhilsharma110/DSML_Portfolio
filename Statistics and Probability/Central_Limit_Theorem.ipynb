# The Central Limit Theorem

The central limit theorem (CLT) is a statistical theory that states that the distribution of sample means will approximate a normal distribution as the sample size increases. This is true regardless of the population's distribution.

The CLT states that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough. This is true regardless of whether the population has a uniform, exponential, normal, Poisson, binomial, or any other distribution.

## Load dependencies

```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Simulating a uniformaly distributed population

Generate 1 samples (uniform[0,1])

```
data = np.random.uniform(size=1)
data
```

Generate 100 samples (uniform[0,1]) and plot the distribution using histplot

```
data = np.random.uniform(size=100)
sns.histplot(data=data, color='orange')
plt.title("100 uniformly distributed samples")
plt.xlabel("Values")
plt.show()
```
